## 2.12 Файловый объект

**.read()**

```python
name='name.txt'
writeln=['Hallo!\n', 'How are u?\n', 'How was ur day?\n']

file=open(name, 'w', encoding='utf-8')

file.writelines(writeln)
file.write('I am OK!\n')
file.close()

file=open(name, 'r', encoding='utf-8')
readln=file.read()
print(readln)
file.close()

```

```python
file=open(name, 'r', encoding='utf-8')
line=file.readline()
while line:
    print(line)
    line=file.readline()
file.close()
```

**with**

```python
with open(name, 'r', encoding='utf-8') as file:
    for line in file:
        print(line)
```

### 2. Почему результат выражения 0.1 + 0.2 не равен 0.3?

float в CPython является аналогом 64-битного double в C/C++ . Он не может точно представить большинство десятичных дробей в двоичной системе, что приводит к ошибкам округления. Вместо точного значения 0.3 мы получаем очень близкое, но не равное ему число.

```python
x = 0.1
y = 0.2
z = x + y
print(f"x + y = {z}")
print(f"Требуемое значение: 0.3")
print(f"Проверка на равенство: {z == 0.3}")
```

```python

print(f"Значение 0.1 + 0.2 с высокой точностью: {z:.20f}")
print(f"Значение 0.3 с высокой точностью: {0.3:.20f}")
```

```python
import math
if math.isclose(z, 0.3, rel_tol=1e-5, abs_tol=1e-8):
    print("math.isclose: числа равны с заданной относительной и абсолютной точностью.")
```

### 5. Почему ключом в словаре может быть кортеж, но не может быть список?

Ключи в словаре должны быть объектами неизменяемых типов . Словари внутри реализованы как хеш-таблицы. Ключ преобразуется в хеш-значение, которое используется для быстрого поиска. Если ключ изменится после того, как был помещен в словарь, его хеш также изменится, и словарь больше не сможет его найти.

Кортеж (tuple) — неизменяемый тип. После создания его содержимое нельзя изменить, поэтому его хеш всегда будет постоянным.

Список (list) — изменяемый тип. Мы можем добавлять, удалять или изменять его элементы, что приведет к изменению его хеша, что сделало бы словарь неработоспособным.

```python
my_dict = {}
coord = (10, 20)
my_dict[coord] = "Точка A"
print(f"Словарь с кортежем в качестве ключа: {my_dict}")
print(f"Значение по ключу (10,20): {my_dict[(10, 20)]}")
```

```python
my_dict1 = {}
try:
    my_dict[[10, 20]] = "Точка B" # Прямая попытка тоже вызовет ошибку
except TypeError as e:
    print(f"Ошибка при попытке использовать список как ключ: {e}")
```

```python
try:
    bad_tuple = (1, 2, [3, 4])
    my_dict[bad_tuple] = "Ошибка"
except TypeError as e:
    print(f"Ошибка для кортежа со списком: {e}")
```

### 6. Если кортеж неизменяем, почему выражение t = ([1, 2], 3); t[0].append(3) работает и изменяет t?

Неизменяемость кортежа относится только к его верхнему уровню, то есть к самому контейнеру и набору содержащихся в нем ссылок . Кортеж t хранит две ссылки: на список [1, 2] и на целое число 3.

Нельзя изменить сам набор ссылок кортежа: например, попытаться присвоить t[0] = something_else.

Однако, если ссылка указывает на объект изменяемого типа (как список), сам этот объект можно изменить.

```python
t = ([1, 2], 3)
print(f"Исходный кортеж: {t}")
print(f"id(t): {id(t)}")
print(f"id(t[0]): {id(t[0])} (ссылка на список)")
```

```python
try:
    t[0] = [5, 6]
except TypeError as e:
    print(f"Ошибка при попытке изменить ссылку в кортеже: {e}")
```

```python
t[0].append(3)
print(f"Кортеж после изменения списка: {t}")
print(f"id(t): {id(t)} (тот же кортеж)")
print(f"id(t[0]): {id(t[0])} (та же ссылка, но список внутри изменился)")
```
