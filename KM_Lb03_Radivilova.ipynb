**Определение функции**

```python
import numpy as np
import matplotlib.pyplot as plt

def y(x):
    return np.sin(x)**5
```

**Создание переменных**

```python
x_min=1.0
x_max=5.0
step=0.05
```

```python
x_array=np.arange(x_min, x_max, step)
y_array=y(x_array)
```

**Создание матрицы из двух массивов**

```python
matrix=np.transpose([x_array, y_array])
```

**Создание  графического окна**

```python
fig1=plt.figure()
```

**Оси графика**

```python
y_min=np.min(y_array)
y_max=np.max(y_array)

ax1=plt.axes()
plt.axis([x_min-0.1, x_max+0.1, y_min-0.1, y_max+0.1])
```
